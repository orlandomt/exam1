1. Escribir una función que reciba un arreglo unidimensional numpy de numeros enteros y devuelva un valor `True` si todos los elementos son unicos o un valor `False` si alguno de los elementos se repite.

> Por ejemplo:

 
> Input 1: [51, 42, 32, 47, 35, 78]

> Output 1: True

> Input 2: [45, **35**, 78, **35**, 10, 98]

> Output 2: False



````
# Escribir la solución en la siguiente linea
import numpy as np

def validar_duplicado(array):
  element, count = np.unique(array, return_counts=True)

  for i in np.asarray(count):
    if i > 1:
      return False
  return True

# Ejemplo 1
array1 = np.array([51, 42, 32, 47, 35, 78])
x = validar_duplicado(array1)
print("Ejemplo 1: Array {0} es {1}".format(array1, x))

# Ejemplo 2
array2 = np.array([45, 35, 78, 35, 10, 98])
y = validar_duplicado(array2)
print("Ejemplo 2: Array {0} es {1}".format(array2, y))
````


5. Graficar una función `sin(x)` en el rango de $-\pi$ a $\pi$ donde se debe señalar mediante una flecha el punto minimo y maximo de la curva indicando ademas sus coordenadas.



````
# Escribir la solución en la siguiente linea
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-1*np.pi,np.pi,0.1)
y = x*np.sin(x)

ymax = max(y)
xmax = max(x)

ymin = min(y)
xmin = min(x)

print("Coordenadas punto máximo")
print("x: {0}".format(xmax))
print("y: {0}".format(ymax))
print("")
print("Coordenadas punto mínimo")
print("x: {0}".format(xmin))
print("y: {0}".format(ymin))
print("")

plt.plot(x,y)
plt.ylim(-1, 5)
plt.xlim(-4, 6)
plt.annotate('max ({0},{1})'.format(round(xmax,2), round(ymax,2)), xy=(xmax, ymax), xytext=(3, 4),
             arrowprops=dict(facecolor='black', shrink=0.05),)
plt.annotate('max ({0},{1})'.format(round(xmin,2), round(ymin,5)), xy=(xmin, ymin), xytext=(-3, 4),
             arrowprops=dict(facecolor='black', shrink=0.05),)
plt.show()
````

6. Escribir un programa que realice la multiplicación elemento por elemento en la **GPU** de 4 matrices de tamaño (M, N) donde M y N son numeros enteros. Además, preparar un ejemplo para mostrar su funcionamiento.


````
# Escribir la solución en la siguiente linea
from numba import cuda
import timeit
import numpy as np
from IPython.display import clear_output
import matplotlib.pyplot as plt

@cuda.jit
def matmul(M1, N1, R1, M2, N2, R2, R3):
  i, j = cuda.grid(2)

  if i < R.shape[0] and j < R.shape[1]:
    R1[i, j] = M1[i, j] * N1[i, j]
    R2[i, j] = M2[i, j] * N2[i, j]
    R3[i, j] = R1[i, j] * R2[i, j]
````

````
dim = 1000
M1 = 2*np.ones([dim, 2*dim])
N1 = 3*np.ones([dim, 2*dim])
R1 = np.zeros([dim, 2*dim])

M1_arr = cuda.to_device(M1)
N1_arr = cuda.to_device(N1)
R1_arr = cuda.to_device(R1)

M2 = 2*np.ones([dim, 2*dim])
N2 = 3*np.ones([dim, 2*dim])
R2 = np.zeros([dim, 2*dim])

M2_arr = cuda.to_device(M2)
N2_arr = cuda.to_device(N2)
R2_arr = cuda.to_device(R2)

R3 = np.zeros([dim, 2*dim])
R3_arr = cuda.to_device(R3)

threadsperblock = (16, 16)
blockspergrid_x = np.int(R1.shape[0] / threadsperblock[0]) + 1
blockspergrid_y = np.int(R1.shape[1] / threadsperblock[1]) + 1
blockspergrid = (blockspergrid_x, blockspergrid_y)

matmul[blockspergrid, threadsperblock](M1_arr, N1_arr, R1_arr, M2_arr, N2_arr, R2_arr, R3_arr)

R1_arr.copy_to_host(R1)
R2_arr.copy_to_host(R2)
R3_arr.copy_to_host(R3)

R3
````
